### Подготовим данные в массиве train

```python
train.describe()
```

Переведем id в строковый тип

```python
train['Id'] = train['Id'].astype(str)
train['Id'].dtype
```

Работаем с данными по комнатам. Очень большое количество комнат в небольшой плашади считаем неверными данными и нудевой количество комнат относим к студияям. Большое количество комнат меняем на значение 1, а нулевое на медиану.

```python
train['Rooms'].value_counts()
```

```python
train.loc[(train['Rooms'] > 5) | (train['Rooms'] == 0)]
```

```python
train['Rooms_bad_numbers'] = 0
train.loc[(train['Rooms'] > 5) | (train['Rooms'] == 0), 'Rooms_bad_numbers'] = 1
```

```python
train.loc[train['Rooms'] == 0, 'Rooms'] = 1
train.loc[train['Rooms'] > 5, 'Rooms'] = train['Rooms'].median()
train['Rooms'].value_counts()
```

Теперь надо проверить нет ли в данных ошибки по площади. Проверим не превышает ли жилая площадь или площадь кухни общую площадь квартиры. Так же проверим нет ли нудевых или отрицательных площадей. Нашли что есть квартиры, где жилая площадь больше основной площади. Введем показатель плохих данных по площади и приравняем такую площадь к сумме площадей жилой и кухни.

```python
train.loc[(train['Square'] < train['LifeSquare']) \
          | (train['Square'] < (train['LifeSquare'] + train['KitchenSquare']))]
```

```python
train['Square_bad'] = 0
train.loc[(train['Square'] < train['LifeSquare']) \
          | (train['Square'] < (train['LifeSquare'] + train['KitchenSquare'])), 'Square_bad'] = 1
train['Square_bad'].value_counts()
```

```python
train.loc[train['Square'] < train['LifeSquare'], 'Square'] = train['LifeSquare'] + train['KitchenSquare']
train.loc[train['Square'] < (train['LifeSquare'] + train['KitchenSquare']), 'Square'] \
= train['LifeSquare'] + train['KitchenSquare']
```

Посмотрим на данные по этажности. Нулевой этажности быть не может. Еще проверим на соответсвие этажа квартиры и этажности дома. Все эти ошибки в данных заменим на медиану этажности дома.

```python
train['HouseFloor'] = train['HouseFloor'].astype(int)
train['HouseFloor'].dtype
```

```python
train['HouseFloor'].value_counts()
```

```python
train['HouseFloor_bad'] = 0
train.loc[(train['HouseFloor'] == 0) | (train['Floor'] > train['HouseFloor']), 'HouseFloor_bad'] = 1
train['HouseFloor_bad'].value_counts()
```

```python
train.loc[(train['HouseFloor'] == 0) | (train['Floor'] > train['HouseFloor']), 'HouseFloor'] \
        = train['HouseFloor'].median()
```

Посмотрим на данные по году постройки дома. Год не может быть больше 2021.

```python
train['HouseYear'].sort_values()
```

```python
train['HouseYear_bad'] = 0
train.loc[train['HouseYear'] > 2020, 'HouseYear_bad'] = 1
train.loc[train['HouseYear'] > 2020, 'HouseYear'] = 2020
```

Жилую площадь рассчитываем как разницуу общей площади квартиры и площади кухни.

```python
train['LifeSquare_null'] = train['LifeSquare'].isna() * 1
train.loc[train['LifeSquare'].isna(), 'LifeSquare'] = train['Square'] - train['KitchenSquare']
```
