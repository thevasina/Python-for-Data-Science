### Класс для обработки данных

```python
class DataPreprocessing:
    def __init__(self):
        self.median = None
        
    def fit(self, X):
        self.median = X.median()
        self.kitchen_square_quantile = X['KitchenSquare'].quantile(.975)
          
    def transform(self, X):
        
        #Комнаты
        X['Rooms_bad_numbers'] = 0
        X.loc[(X['Rooms'] > 6) | (X['Rooms'] == 0), 'Rooms_bad_numbers'] = 1
        X.loc[X['Rooms'] == 0, 'Rooms'] = 1
        X.loc[X['Rooms'] > 5, 'Rooms'] = self.median['Rooms']
        
        #Площадь квартиры
        X['Square_bad'] = 0
        X.loc[(X['Square'] < X['LifeSquare']) | (X['Square'] \
                                < (X['LifeSquare'] + X['KitchenSquare'])), 'Square_bad'] = 1
        X.loc[X['Square'] < X['LifeSquare'], 'Square'] = X['LifeSquare'] + X['KitchenSquare']
        X.loc[X['Square'] < (X['LifeSquare'] + X['KitchenSquare']), 'Square'] \
                                = X['LifeSquare'] + X['KitchenSquare']
        
        #Этажи
        X['HouseFloor_bad'] = 0
        X.loc[(X['HouseFloor'] == 0) | (X['Floor'] > X['HouseFloor']), 'HouseFloor_bad'] = 1
        X.loc[(X['HouseFloor'] == 0) | (X['Floor'] > X['HouseFloor']), 'HouseFloor'] \
        = self.median['HouseFloor']
        
        #Год постройки
        current_year = datetime.now().year  
        X['HouseYear_bad'] = 0
        X.loc[X['HouseYear'] > current_year, 'HouseYear_bad'] = 1
        X.loc[X['HouseYear'] > current_year, 'HouseYear'] = current_year
        
        #Заполнение пропусков в площади
        X['LifeSquare_null'] = X['LifeSquare'].isna() * 1
        X.loc[X['LifeSquare'].isna(), 'LifeSquare'] = X['Square'] - X['KitchenSquare']
        
        #Удаление признака Healthcare_1
        X.drop('Healthcare_1',axis = 1, inplace = True)
        
        #Обработка выбросов в площадях кухни
        condition = (X['KitchenSquare'].isna()) \
                    | (X['KitchenSquare'] > self.kitchen_square_quantile)
        
        X.loc[condition, 'KitchenSquare'] = self.median['KitchenSquare']

        X.loc[X['KitchenSquare'] < 3, 'KitchenSquare'] = 3
        
        X.fillna(self.median, inplace=True)
               
        return X
```

### Класс для генерации новых признаков

```python
class FeatureGenetator():
    def __init__(self):
        self.str_to_numbers = None
        self.med_price_by_district = None
        self.sq_metr_price = None
        self.df_for_merge = None
        
    def fit(self, X, y=None):
        X = X.copy()
        self.str_to_numbers = {'A': 0, 'B': 1}
        
        df = X.copy()
        
        if y is not None:
            df['Price'] = y.values
            
            self.med_price_by_district = df.groupby(['DistrictId', 'Rooms'], as_index=False).agg({'Price':'median'})\
                                            .rename(columns={'Price':'MedPriceByDistrict'})
            
            self.med_price_by_district_median = self.med_price_by_district['MedPriceByDistrict'].median()
        
        if y is not None:
            df['Price'] = y.values
            
            self.df_for_merge = df[['Square', 'Rooms']]
            self.df_for_merge = self.df_for_merge.assign(sq_price=df['Price'] / df['Square'])
            
            
    def transform(self, X):
        
        # Категория возраста дома
        X['year_cat'] = 0
        X.loc[train['HouseYear'] <= 1941, 'year_cat'] = 1
        X.loc[(train['HouseYear'] > 1941) & (train['HouseYear'] <= 1945), 'year_cat'] = 2
        X.loc[(train['HouseYear'] > 1945) & (train['HouseYear'] <= 1980), 'year_cat'] = 3
        X.loc[(train['HouseYear'] > 1980) & (train['HouseYear'] <= 2000), 'year_cat'] = 4
        X.loc[(train['HouseYear'] > 2000) & (train['HouseYear'] <= 2010), 'year_cat'] = 5
        X.loc[(train['HouseYear'] > 2010), 'year_cat'] = 6
        
        #Стоимость квадратного метра
        #X['sq_metr_price'] = 0
        #X['sq_metr_price'] = dr / X['Squre']
    
        #Перевод строковых данных в числовые
        X['Ecology_2'] = X['Ecology_2'].replace(str_to_numbers)
        X['Ecology_3'] = X['Ecology_3'].replace(str_to_numbers)
        X['Shops_2'] = X['Shops_2'].replace(str_to_numbers)
        
        if self.med_price_by_district is not None:
            X = X.merge(self.med_price_by_district, on=['DistrictId', 'Rooms'], how='left')
            X.fillna(self.med_price_by_district_median, inplace=True)
            
        if self.df_for_merge is not None:
            X = X.merge(self.df_for_merge, on=['Square', 'Rooms'], how='left')
       
       
        
        return X
```
